```python
from matplotlib import image
from matplotlib import pyplot as pp
import numpy as np

img1=image.imread("ATM.jpg")  # Call Atm Image
pp.imshow(img1)
pp.axis("off")                  # Show Atm Image
pp.show()

options = ["Transfer money","Account balance", "Change Password", "Deposit", "Withdrawal"]
for o in options:
  print("*", o.title(), " \t*")
print("Welcome to Diamond Bank ATM, Kindly insert your card")
#3 users # list of dictionaries
user1 = {"name": "Joy", "account number": "12345", "password": "xyz", "account balance": 20000} 
user2 = {"name": "Jane", "account number": "67890", "password": "abc", "account balance": 3000}
user3 = {"name": "Phil", "account number": "73476", "password": "hjy", "account balance": 5000}
  
x=input("Enter your name")    # user input function
y=input("Enter your account number")
z=input("Enter your password")  

if x == user1.get("name"): 
  if y == user1.get("account number"):
    if z == user1.get("password"):
      print("Login Successful", "your account balance is", user1.get("account balance"))
      print(["Transfer money","Account balance", "Change Password", "Deposit", "Withdrawal"])
      a=True
    while a==True: #while loop
      a = input("What transaction would you like to make? Kindly select from the options.")
      if a == "Deposit":      # conditional statment
        user_depo=int(input("How much would you like to deposit?")) # money to be deposited
        user1_curr=user1.get("account balance")
        user1_curr = int(user1_curr) + int(user_depo)
        print("Your current balance is", user1_curr)               # money deposited
        user1["account balance"]=user1_curr                        # new current balance
        a=True
      if a == "Withdrawal":
        user_with=int(input("How much would you like to withdraw?"))    # money to be withdrawn
        user1_curr=user1.get("account balance")
        user1_curr = int(user1_curr) - int(user_with)                   # money withdrawn
        print("Your current balance is", user1_curr)               
        user1["account balance"]=user1_curr                            # new current balance 
        a=True
      if a == "Transfer":
        user_trans=int(input("How much would you like to transfer?"))   #money to be transferred
        user1_curr=user1.get("account balance")
        user1_curr = int(user1_curr) - int(user_trans)                  #money transferred
        print("Your current balance is", user1_curr)
        user1["account balance"]=user1_curr                             #new current balance
        a=True
      if a == "Change password":
        user_pass=str(input("Kindly insert your current password?"))   # password change
        if user1.get("password") == user_pass:
         new_pass=input("Enter a new password")                         # new password 
         user1["password"] = new_pass  
         print("Your new password is", user1["password"])               #new current password
        a=True
      
      if a=="quit":
        print("Thank you for banking with Diamond bank")
    
if x == user2.get("name"):
  if y == user2.get("account number"):
    if z == user2.get("password"):
      print("Login Successful", "your account balance is", user2.get("account balance"))
      print(["Transfer money","Account balance", "Change Password", "Deposit", "Withdrawal"])
      a=True
    while a==True:
      a = input("What transaction would you like to make? Kindly select from the options.")
      if a == "Deposit":
        user_depo=int(input("How much would you like to deposit?"))
        user2_curr=user2.get("account balance")
        user2_curr = int(user2_curr) + int(user_depo)
        print("Your current balance is", user2_curr)
        user2["account balance"]=user2_curr
        a=True
      if a == "Withdrawal":
        user_with=int(input("How much would you like to withdraw?"))
        user2_curr=user2.get("account balance")
        user2_curr = int(user2_curr) - int(user_with)
        print("Your current balance is", user2_curr)
        user2["account balance"]=user2_curr
        a=True
      if a == "Transfer":
        user_trans=int(input("How much would you like to transfer?"))
        user2_curr=user2.get("account balance")
        user2_curr = int(user2_curr) - int(user_trans)
        print("Your current balance is", user2_curr)
        user2["account balance"]=user2_curr
        a=True
      if a == "Change password":
        user_pass=str(input("Kindly insert your current password?"))
        if user2.get("password") == user_pass:
         new_pass=input("Enter a new password")
         user2["password"] = new_pass  
         print("Your new password is", user2["password"])
         a=True
      
      if a=="quit":
        print("Thank you for banking with Diamond bank")
    
if x == user3.get("name"):
  if y == user3.get("account number"):
    if z == user3.get("password"):
      print("Login Successful", "your account balance is", user3.get("account balance"))
      print(["Transfer money","Account balance", "Change Password", "Deposit", "Withdrawal"])
      a=True
    while a==True:
      a = input("What transaction would you like to make? Kindly select from the options.")
    if a == "Deposit":
        user_depo=int(input("How much would you like to deposit?"))
        user3_curr=user3.get("account balance")
        user3_curr = int(user3_curr) + int(user_depo)
        print("Your current balance is", user3_curr)
        user3["account balance"]=user3_curr
        a=True
    if a == "Withdrawal":
        user_with=int(input("How much would you like to withdraw?"))
        user3_curr=user3.get("account balance")
        user3_curr = int(user3_curr) - int(user_with)
        print("Your current balance is", user3_curr)
        user3["account balance"]=user3_curr
        a=True
    if a == "Transfer":
        user_trans=int(input("How much would you like to transfer?"))
        user3_curr=user3.get("account balance")
        user3_curr = int(user3_curr) - int(user_trans)
        print("Your current balance is", user3_curr)
        user3["account balance"]=user3_curr
        a=True
    if a == "Change password":
        user_pass=str(input("Kindly insert your current password?"))
        if user3.get("password") == user_pass:
         new_pass=input("Enter a new password")
         user3["password"] = new_pass  
         print("Your new password is", user3["password"])
        a=True

    if a=="quit":
      print("Thank you for banking with Diamond bank")
    
```
